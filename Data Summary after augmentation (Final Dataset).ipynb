```python
import os

def count_files_in_subdirectories(directory):
  """Counts the total number of files in each subdirectory within a directory.

  Args:
    directory: The path to the directory containing subdirectories.

  Returns:
    A dictionary where keys are subdirectory names and values are file counts,
    sorted in descending order.
  """
  file_counts = {}
  for root, subdirectories, files in os.walk(directory):
    # Skip the top-level directory (optional)
    if root == directory:
      continue
    # Get subdirectory name
    subdirectory_name = os.path.basename(root)
    # Count files in the subdirectory
    file_count = len(files)
    file_counts[subdirectory_name] = file_count
  # Sort dictionary by values (file counts) in descending order
  return dict(sorted(file_counts.items(), key=lambda item: item[1], reverse=True))

```

## Train

```python
# Example usage (replace 'path/to/your/directory' with your actual path)
master_directory = 'C:/Users/Hp/Desktop/IIT Research Docs/Movie Labeling/Augmented image_two_three/Train_TwoThree_Aug'
import os
import pandas as pd
# Function to count the number of files in a directory
def count_files(directory):
    return len([name for name in os.listdir(directory) if os.path.isfile(os.path.join(directory, name))])

# Initialize an empty list to store directory names and counts
data = []

# Iterate through each item in the master directory
for item in os.listdir(master_directory):
    item_path = os.path.join(master_directory, item)
    if os.path.isdir(item_path):
        num_files = count_files(item_path)
        data.append({'Genre': item, 'Number_of_Files': num_files})

# Create a DataFrame from the list of dictionaries
train_df = pd.DataFrame(data)
```

```python
train_df.sort_values(by="Number_of_Files",ascending=False, ignore_index=True,inplace=True)
```
