```python
import os

def list_files_in_directory(directory):
    file_list = []
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            file_name_without_extension = os.path.splitext(filename)[0]
            file_list.append(int(file_name_without_extension))
    return file_list

# Replace 'directory_path' with the path of your directory
directory_path = 'C:/Users/Hp/Desktop/IIT Research Docs/Movie Labeling/images_new'
files = list_files_in_directory(directory_path)
```

```python
gen = filtered_df["Genre"]
```

```python
# Split the 'series' column by '|' delimiter and expand it into separate columns
split_series = gen.str.split('|', expand=True)

# Rename the columns to series_1, series_2, ...
split_series.columns = [f'series_{i+1}' for i in range(split_series.shape[1])]

# Concatenate the original DataFrame with the split series DataFrame
result_df = pd.concat([filtered_df, split_series], axis=1)

print(result_df)
```
