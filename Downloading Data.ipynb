```python
import pandas as pd
import numpy as np
```

```python
genres = pd.read_csv("MovieGenre.csv", encoding="latin1")
```

```python
id_and_link = genres[["imdbId","Poster"]].iloc[:10250,:]
```

```python
id_and_link.dropna(inplace=True)
```

```python
import requests
import os

def download_image(row):
  name = row['imdbId']
  link = row['Poster']
  directory = 'C:/Users/Hp/Desktop/IIT Research Docs/Movie Labeling/images_new'
  response = requests.get(link)
  if response.status_code == 200:
    if not os.path.exists(directory):
      os.makedirs(directory)
    with open(f"{directory}/{name}.jpg", 'wb') as f:  # Corrected line with closing parenthesis
      f.write(response.content)
      print(f"Downloaded {name}.jpg")
  else:
    print(f"Failed to download {name}.jpg")


# Download images
id_and_link.apply(download_image, axis=1)
```
