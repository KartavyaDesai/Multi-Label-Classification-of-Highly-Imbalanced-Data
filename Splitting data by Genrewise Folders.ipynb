```python
import pandas as pd
import numpy as np
```

```python
df_tr = pd.read_csv("train.csv", index_col="Unnamed: 0").reset_index()
```

```python
df_tr
```

```python
filenames = []
for item in df_tr["imdbId"]:
    filenames.append(str(item)+".jpg")
filename = pd.Series(filenames)
```

```python
df_tr["filename"] = filename
```

```python
df_tr.drop(columns=["index","imdbId","Title","IMDB Score"],inplace=True)
```

```python
df_tr.head()
```

```python
# Splitting the 'Genre' column and expanding it into separate columns
genres = df_tr['Genre'].str.split('|', expand=True)

# Naming the new columns
genres.columns = [f'Genre_{i+1}' for i in range(genres.shape[1])]

# Concatenating the new columns with the original DataFrame
df_tr = pd.concat([df_tr, genres], axis=1)

# Dropping the original 'Genre' column if needed
df_tr.drop(columns=['Genre'], inplace=True)
```

```python
df_tr
```

```python
import shutil
import os


# Path to the folder containing images
image_folder = 'C:/Users/Hp/Desktop/IIT Research Docs/Movie Labeling/train_imgs'

# Iterate through the DataFrame
for index, row in df_tr.iterrows():
    image_name = row['filename']
    for genre_col in ['Genre_1', 'Genre_2', 'Genre_3']:
        genre = row[genre_col]
        if genre:
            # Create the genre folder if it doesn't exist
            genre_folder = os.path.join(image_folder, genre)
            if not os.path.exists(genre_folder):
                os.makedirs(genre_folder)
            # Copy the image into the genre folder
            shutil.copy(os.path.join(image_folder, image_name), genre_folder)
```
